```r
# Load readxl package
library("readxl")
library("ggplot2")
library("tidyr")
library("dplyr")

```

```r
df = read_excel("/kaggle/input/carmileage/carmileage.xlsx")
cardata=df[,-1]
```

```r
head(df)
```

```r
summary(df)
sapply(cardata,sd)
sapply(cardata,hist)
```

```r
#Visualizing 5 number summary
boxplot(df$VOL, main = "Volume (Cubic feet) of cab space")
boxplot(df$HP, main = "Engine Horsepower")
boxplot(df$MPG, main = "Miles Per Gallon")
boxplot(df$WT, main = "Weight (lbs)")
boxplot(df$SP, main = "Top Speed")
```

```r
df$VOL_cat <- cut(df$VOL, breaks=c(49, 89.5, 101, 113, 160))
df$HP_cat <- cut(df$HP, breaks=c(48, 84, 99, 140, 322))
df$SP_cat <- cut(df$SP, breaks=c(89, 105, 109, 114.8, 165))
df$WT_cat <- cut(df$WT, breaks=c(17, 25, 30, 35, 55))
df$MPG_cat <- cut(df$MPG, breaks=c(13, 27.77, 32.45, 39.30, 65.40))

agg_VOL = aggregate(df,
                by = list(df$VOL_cat),
                FUN = mean)

ggplot(agg_VOL, 
       aes(x = Group.1, 
           y = MPG)) +
  geom_bar(stat = "identity")

agg_HP = aggregate(df,
                    by = list(df$HP_cat),
                   FUN = mean)

ggplot(agg_HP, 
       aes(x = Group.1, 
           y = MPG)) +
  geom_bar(stat = "identity")

agg_SP = aggregate(df,
                    by = list(df$SP_cat),
                    FUN = mean)
ggplot(agg_SP, 
       aes(x = Group.1, 
           y = MPG)) +
  geom_bar(stat = "identity")

agg_WT = aggregate(df,
                    by = list(df$WT_cat),
                    FUN = mean)
ggplot(agg_WT, 
       aes(x = Group.1, 
           y = MPG)) +
  geom_bar(stat = "identity")

agg_MPG = aggregate(df,
                    by = list(df$MPG_cat),
                    FUN = mean)
```

```r
#Vol Vs MPG
cor.test(df$MPG, df$VOL, method = "pearson")

attach(df)
plot(VOL, MPG, main="Scatterplot of Volume and MPG", 
     xlab="Volume of cab space ", ylab="Miles Per Gallon ", pch=19)
abline(lm(MPG ~ VOL), col = 4, lwd = 3)

df %>% 
  ggplot(aes(x=VOL, y=MPG)) +
  geom_jitter(alpha=.5) +
  geom_rug(position="jitter", size=.08) +
  geom_smooth(size=.6) +
  labs(title="Volume and MPG", x="Volume of cab space", y="Miles Per Gallon") +
  theme(plot.title = element_text(hjust = 0.5, size=10, face = "bold"))

#HP Vs MPG
cor.test(df$MPG, df$HP, method = "pearson")

plot(HP, MPG, main="Scatterplot of Horsepower and MPG", 
     xlab="Horsepower ", ylab="Miles Per Gallon ", pch=19)
abline(lm(MPG ~ HP), col = 4, lwd = 3)

df %>% 
  ggplot(aes(x=HP, y=MPG)) +
  geom_jitter(alpha=.5) +
  geom_rug(position="jitter", size=.08) +
  geom_smooth(size=.6) +
  labs(title="Horsepower and MPG", x="Horsepower", y="Miles Per Gallon") +
  theme(plot.title = element_text(hjust = 0.5, size=10, face = "bold"))

#WT Vs MPG
cor.test(df$MPG, df$WT, method = "pearson")

plot(WT, MPG, main="Scatterplot of Weight and MPG", 
     xlab="Weight ", ylab="Miles Per Gallon ", pch=19)
abline(lm(MPG ~ WT), col = 4, lwd = 3)

df %>% 
  ggplot(aes(x=WT, y=MPG)) +
  geom_jitter(alpha=.5) +
  geom_rug(position="jitter", size=.08) +
  geom_smooth(size=.6) +
  labs(title="Weight and MPG", x="Weight of cab", y="Miles Per Gallon") +
  theme(plot.title = element_text(hjust = 0.5, size=10, face = "bold"))

#SP Vs MPG
cor.test(df$MPG, df$SP, method = "pearson")

plot(SP, MPG, main="Scatterplot of Top Speed and MPG", 
     xlab="Top Speed ", ylab="Miles Per Gallon ", pch=19)
abline(lm(MPG ~ SP), col = 4, lwd = 3)

df %>% 
  ggplot(aes(x=SP, y=MPG)) +
  geom_jitter(alpha=.5) +
  geom_rug(position="jitter", size=.08) +
  geom_smooth(size=.6) +
  labs(title="Top Speed and MPG", x="Top Speed of cab", y="Miles Per Gallon") +
  theme(plot.title = element_text(hjust = 0.5, size=10, face = "bold"))
```

```r
# Multiple Linear Regression Example 
fit <- lm(MPG ~ VOL + HP + WT + SP, data=df)
summary(fit)

#Residuals
res <- resid(fit)
#produce residual vs. fitted plot
plot(fitted(fit), res)
#add a horizontal line at 0 
abline(0,0)

anova(fit)
```

```r
#: Regression of MPG on all the explanatory variables
car.lm=lm(MPG~HP+VOL+SP+WT,data=cardata)
summary(car.lm)


#Refine the model using only HP,VOL  as explanatory variables
car.lm2=lm(MPG~HP+VOL,data=cardata)
summary(car.lm2)

#Examine Residuals:
#Plot the residuals vs. the predicted values y_hat to see if there is evidence of a curved (rather than linear)relation
plot(fitted(car.lm2),residuals(car.lm2))
abline(0,0)

plot(fitted(car.lm),residuals(car.lm))
abline(0,0)

#Plot the residuals vs. each of the explanatory variables to check if the variances of the residuals are constant
plot(HP,residuals(car.lm2),xlab="HP",ylab="Residuals")
abline(0,0)
plot(VOL,residuals(car.lm),xlab="VOL",ylab="Residuals")
abline(0,0)
plot(SP,residuals(car.lm),xlab="SP",ylab="Residuals")
abline(0,0)
plot(WT,residuals(car.lm),xlab="WT",ylab="Residuals")
abline(0,0)

#nomal quantile plot for residuals to check normality of residuals
qqnorm(residuals(car.lm2),xlab="Normal Score", ylab="Residuals")
qqline(residuals(car.lm2))
```
